**Uses a label encoder to transform the two target strings into binary**

```python
label_encoder = LabelEncoder()
df['Class'] = label_encoder.fit_transform(df['Class'])

df.head
```

The class column now has 1s and 0s

**Creates a boxplot with whiskers for all the data to check the range and see if there are any outliers**

```python
numerical_cols = df.select_dtypes(include=["float64", "int64"]).columns.tolist()
figure, axes = plt.subplots(nrows=2, ncols=4, figsize=(20,44), constrained_layout = True)

i = j = 0

for column in numerical_cols:
    sns.boxplot(y=df[column], ax=axes[i, j])

    if j == 3:
        j = 0
        i += 1
    else:
        j += 1
    
```

**Creates a correlation matrix to check how different variables influence each other**

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, vmin=-1, vmax=1)
plt.show
```

Most variables influence each other. Extent and eccentricity do not. I will use area for the variable out of the influencial ones

**Replotted correlation matrix without values that influence one another**

```python
columns_to_drop = ["MajorAxisLength", "MinorAxisLength", "ConvexArea", "Perimeter", "Class"]
df_processed_matrix = df.drop(columns_to_drop, axis=1)

corr_matrix = df_processed_matrix.corr()
sns.heatmap(corr_matrix, annot=True, vmin=-1, vmax=1)
plt.show
```

Columns still influence each other, but not as much as the others

**Graphs to show how the spread of values for each column for both target values**

```python
numerical_cols = df.select_dtypes(include=["float64", "int64"]).columns.tolist()
figure, axes = plt.subplots(nrows=4, ncols=2, figsize=(20,20), constrained_layout = False)

i = j = 0

for column in numerical_cols:
    sns.kdeplot(data=df, x=df[column], y=df["Class"],fill=True, ax=axes[i, j])

    if j == 1:
        j = 0
        i += 1
    else:
        j += 1
```
